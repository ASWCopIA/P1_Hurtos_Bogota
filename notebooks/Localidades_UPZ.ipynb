```python
#Carga de archivos - localidades upz
xl = pd.ExcelFile('/content/drive/My Drive/P1_Hurtos_Bogota/data/processed/localidades_y_barrios_con_upz_v2.xlsx')
file_sheets = len(xl.sheet_names)
file_sheets
localidad_upz_array= []
for i in np.arange(file_sheets):
    temp_df = pd.read_excel(xl,i)
    temp_df['localidad_num'] = i+1
    temp_df['nombre_localidad'] = xl.sheet_names[i]
    localidad_upz_array.append(temp_df)
    

final = pd.concat(localidad_upz_array)
final.head()
```

```python
columns_names = ['Upz', 'Barrio', 'Num', 'Localidad_num', 'Nombre_localidad']
final.columns = columns_names
final.head()
```

```python
hurtos_df = pd.read_excel('/content/drive/My Drive/hurtos_bogota_2010-2019.xlsx')
hurtos_df
```

```python
resultado = pd.merge(hurtos_df, final, left_on = 'Barrio', right_on = 'Barrio')
```
