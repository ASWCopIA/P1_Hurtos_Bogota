```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
#Carga de archivos - localidades y barrios
local_barrios_df = pd.read_excel('../data/processed/localidades_y_barrios.xlsx')
local_barrios_df.head()
```

```python
#Carga de archivos - localidades upz
xl = pd.ExcelFile('../data/processed/localidades_y_barrios_con_upz_v2.xlsx')
file_sheets = len(xl.sheet_names)
file_sheets
localidad_upz_array= []
for i in np.arange(file_sheets):
    temp_df = pd.read_excel(xl,i)
    temp_df['localidad_num'] = i+1
    temp_df['nombre_localidad'] = xl.sheet_names[i]
    localidad_upz_array.append(temp_df)
    
final = pd.concat(localidad_upz_array)
final.head()
```

```python
hurtos_df = pd.read_excel('../data/processed/hurtos_bogota_2010-2019.xlsx')
hurtos_df
```

```python
hurtos_df.head(1)
```

```python
resultado = pd.merge(hurtos_df, final, left_on = 'Barrio', right_on = 'BARRIO')
```
